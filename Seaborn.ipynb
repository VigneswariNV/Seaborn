```python
"""
Seaborn samples
"""

import seaborn as sns
sns.get_dataset_names()   #these datasets are already a part of seaborn so it gets loaded whenever we call it
```

```python
import pandas as pd
from matplotlib import pyplot as plt
df = sns.load_dataset('planets')    #loading the dataet planets from seaborn
df.head()
```

```python
di = sns.load_dataset('iris')    #loading the dataet iris from seaborn
di.head()
```

```python
sns.set_style('white')       #basically the styling. We can add whitegrid,tick and other designs. and also xtick and ytick sizes
sns.displot(di['sepal_width'],kind="ecdf")    #kernal density estimation is made false and it produces a curve shaped continuous wave  displot or distplot
plt.show()
```

```python
sns.set_style('white')       #basically the styling. We can add whitegrid,tick and other designs. and also xtick and ytick sizes
sns.displot(di['sepal_width'],kind="hist")    #kernal density estimation is made false and it produces a curve shaped continuous wave  displot or distplot
plt.show()
```

```python
sns.set_style('darkgrid')
sns.distplot(df['mass'])
plt.show()
```

```python
sns.set_style('white')       #basically the styling. We can add whitegrid,tick and other designs. and also xtick and ytick sizes
sns.displot(di['sepal_width'],kind="kde")    #kernal density estimation is made false and it produces a curve shaped continuous wave  displot or distplot
plt.show()
```

```python
df =  sns.load_dataset('iris')
df.head()
sns.jointplot(x='petal_length',y='sepal_length',data=df,kind="reg")   #uses both scatter and histogram we can even add kde by giving kind="kde"
plt.show()
```

```python
df =  sns.load_dataset('iris')
df.head()
sns.jointplot(x='petal_length',y='sepal_length',data=df,kind='hex')   #kind is hex so the inner boxes look like hexagon
plt.show()
```

```python
sns.set_style("ticks")
sns.pairplot(df,hue='species',diag_kind="hist",kind="scatter",palette="pink")   #palette means color. The diagonal graphs alone are in hist form.
```
